```python
import numpy as np
import matplotlib.pyplot as pit 
```

Definizione della funziona $f(x)$

$f(x) = x^3 + x^2 - 5x + 2$

```python
def f(x):
    return(x**3 + x**2 - 5*x -3)
```

```python
x = 0
f(x)
```

Genero il vettore degli argomenti $x$

```python
x = np.linspace(-30 , 20 , 51)
print(x)
```

Genero il vettore delle immagini $y$

```python
y = f(x)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grafico di f(x)')
plt.grid(True)
plt.ylim(-20, 20)
plt.show()
```
