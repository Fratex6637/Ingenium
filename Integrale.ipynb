```python
import numpy as np
from numpy import sin
from numpy import pi
```

```python
def f(x):
    return (np.sin(x))
```

```python
y = f(2)
print(y)
```

```python
def integrale(f,a,b,n):
    sum_y=0
    x=a
    dx=(b-a)/n
    for i in range (n):
        sum_y=sum_y+f(x)
        x=x+dx
    A=sum_y*dx
    return A
```

```python
Area = integrale(f,0,pi/2,1000)
print('Area sotto la curva =' ,Area)
```

Integrale con trapezzi

```python
def trapezzi(f,a,b,n):
    sum=0
    dx=(b-a)/n
    x=a
    sum=f(x)/2
    x=x+dx
    for i in range (n-1):
        sum=sum+f(x)
        x=x+dx
    sum=sum+f(x)/2
    area = sum*dx
    return (area)
```

```python
area = trapezzi(f,0,pi/2,10)
print('Area con metodo trapezzi =' ,area)
```
